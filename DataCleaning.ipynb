```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

indices = pd.read_csv("climateIndices.csv")
indicesAndIsotopes = pd.read_csv("allValues.csv")
```

```python
#have the same number of data points
print(indices.shape)
print(indicesAndIsotopes.shape)
#checking column names
print(indices.columns.values)
print(indicesAndIsotopes.columns.values)
#stripping columns that repeat
isotopes = indicesAndIsotopes.drop(['SampleName', 'site', 'In', 'Out', 'middate', 'SOI in (monthly)', 'SOI out (monthly)',
 'SOI middate (monthly)', 'SOI average (monthly)', 'PNA in (daily)',
 'PNA out (daily)', 'PNA middate (daily)', 'PNA average (daily)',
 'AO in (daily)', 'AO out (daily)', 'AO middate (daily)', 'AO average (daily)'], axis=1)
```

```python
#let's merge the data so we have actual date values corresponding to the dD and DO values
cleanedData = pd.concat([indices, isotopes], axis=1, join='inner')
print(cleanedData.shape)
cleanedData.to_csv('cleaned.csv')
```
