```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("cleaned.csv")
```

```python
time = pd.to_datetime(data['middate'], format='%m/%d/%Y', errors='coerce')
data.describe()
```

```python
#soi plot (didn't exclude by region)
plt.plot_date(time, data['SOI middate (monthly)'])
plt.show()
```

```python
#plotting pna values
plt.plot_date(time, data['PNA middate (daily)'])
plt.show()
```

```python
#plotting AO values
plt.plot_date(time, data['AO middate (daily)'])
plt.show()
```

```python
#plotting SOI histogram
data.hist(column='SOI middate (monthly)')
plt.show()
```

```python
#plotting PNA histogram
data.hist(column='PNA middate (daily)')
plt.show()
```

```python
#plotting AO histogram
data.hist(column='AO middate (daily)')
plt.show()
```
