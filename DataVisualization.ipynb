```python
#sorting data
dd = data['dD']
do = data['dO']
data = data.drop(['site', 'SampleName', 'In', 'Out', 'middate', 'SOI in (monthly)', 'SOI out (monthly)', 'SOI middate (monthly)', 'PNA in (daily)', 'PNA out (daily)', 'PNA middate (daily)', 'AO in (daily)', 'AO out (daily)', 'AO middate (daily)', 'dD', 'dO'], axis=1)

num_features = data.shape[1]
num_samples = data.shape[0]
print(num_features)
print(num_samples)
```

```python
#initializing weights and bias
weights_init = np.random.uniform(-1, 1, num_features)
bias_init = 0
```

```python
#normalizing
data.apply(pd.to_numeric)
min_max_scaler = preprocessing.MinMaxScaler()
np_scaled = min_max_scaler.fit_transform(data)
data = pd.DataFrame(np_scaled)

# now the values should be normalized:
data.head()

#converting to np arrays
data = data.values
dd = dd.values
do = do.values
```

```python
#training on dd values
regr = linear_model.LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(data, dd, test_size = 0.25, random_state = 5)
regr.fit(X_train, Y_train)
```

```python
plt.scatter(Y_test, regr.predict(X_test))
plt.xlabel("Actual dd values")
plt.ylabel("Predicted dd values")
plt.show()
```

```python
#training on dO values
X_train, X_test, Y_train, Y_test = train_test_split(data, do, test_size = 0.25, random_state = 5)
regr.fit(X_train, Y_train)
```

```python
plt.scatter(Y_test, regr.predict(X_test))
plt.xlabel("Actual dO values")
plt.ylabel("Predicted dO values")
plt.show()
```
